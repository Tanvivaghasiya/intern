```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.linear_model import LogisticRegression

```

```python
nltk.download("punkt")
nltk.download("stopwords")
```

```python
nltk.download('punkt_tab')
```

```python
d=pd.read_csv("spam.csv",encoding="latin-1")
d.head()
```

```python
d.info()
```

```python
d.duplicated().sum()
```

```python
d.drop_duplicates(inplace=True)
```

```python
d.duplicated().sum()
```

```python
d.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],inplace=True)
```

```python
d
```

data preprocessing

```python
def preprocess(text):
  token=word_tokenize(text.lower())
  stop_words=set(stopwords.words("english"))
  token=[word for word in token if word not in stop_words]
  return " ".join(token)
```

```python
d["processedmessage"] = d['v2'].apply(preprocess)
```

```python
d.head()
```

```python
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(d['processedmessage'])
```

```python
y=d['v1'].map({'spam':1,'ham':0})
```

```python
y
```

```python
print(X.shape,y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.6,random_state=12)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

model evaluation

```python
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
confusion=confusion_matrix(y_test,y_pred)
classification=classification_report(y_test,y_pred)
```

```python
print("accuracy score:",accuracy)
print("confusion matrix:",confusion)
print("classificcation report\n",classification)
```

model prediction

```python
input_mail = ["Reminder - Doubt Class Scheduled ( Doubt Buster & Interview Preparation Session @ 8 PM )"]
```

```python
input_data_features = vectorizer.transform(input_mail)
```

```python
prediction = model.predict(input_data_features)
```
