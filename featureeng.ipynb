standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
d=[[2,5],[5,7],[1,4],[7,9]]
```

```python
s = StandardScaler()
a = s.fit(d)
scaled = a.transform(d)

```

```python
print(scaled)
```

min-max

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m=MinMaxScaler()
b=m.fit(d)
scaled=b.transform(d)
```

```python
print(scaled)
```
