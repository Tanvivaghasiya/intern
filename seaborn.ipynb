```python
import seaborn as sns
a=sns.load_dataset('tips')
a.head()
```

```python
b=sns.load_dataset('iris')
b.head()
```

```python
#scatterplot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y="tip",data=a)
plt.show()
```

```python
#lineplot
sns.lineplot(x='total_bill',y='tip',data=a)
plt.show()
```

```python
#barplot
sns.barplot(x='petal_length',y='species',data=b)
plt.show()
```

```python
#histplot
sns.histplot(x='petal_length',data=b)
plt.show()

```

```python
#density plot
sns.kdeplot(x='petal_width',data=b)
plt.show()
```

```python
#boxplot
sns.boxplot(x='petal_length',y='species',data=b)
plt.show()
```

```python
#violinplot
sns.violinplot(x='species',y='petal_width',data=b)
plt.show()
```

```python
#heatmap
a=sns.load_dataset('tips')
a=a.pivot(index='total_bill',columns='tip',values='size')
corr=a.corr()
sns.heatmap(corr)
plt.show()
```
