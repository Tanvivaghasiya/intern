onehot encoding

```python
import pandas as pd
```

```python
city=['Surat','Gandhinagar',"Rajkot","Bharuch"]
```

```python
d=pd.DataFrame(city,columns=['city'])
```

```python
d.head()
```

```python
one=pd.get_dummies(d['city'])
```

```python
print(one)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(d).toarray()
```

```python
print(encoded)
```

```python
d
```
