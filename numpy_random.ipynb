binomial dist

```python
x=random.binomial(n=10,p=0.4,size=9)
print(x)
```

```python
#visulization
sns.displot(random.binomial(n=10,p=0.4,size=9))
plt.show()
```

```python
#diff bet. noraml & binomial

sns.displot(random.normal(loc=5,scale=2,size=300),label="normal")
sns.displot(random.binomial(n=10,p=0.2,size=300),label="binomial")
```

poisson dist

```python
x=random.poisson(lam=4,size=100)
print(x)
```

```python

sns.displot(random.poisson(lam=4,size=100),kde=False)
plt.show()
```

```python
#diff bet. normal and poisson dist

sns.distplot(random.normal(loc=4,scale=5,size=100),hist=True,label="normal")
sns.distplot(random.poisson(lam=4,size=100),hist=True,label="poisson")
plt.show()
```

```python
#diff bet. poisson and binomial

sns.distplot(random.binomial(n=10,p=0.4,size=1000),hist=True,label="binomial")
sns.distplot(random.poisson(lam=4,size=1000),hist=True,label="poisson")
plt.show()
```
