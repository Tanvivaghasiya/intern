```python
#font properties
x=np.array([121,232,111,455,100,200])
y=np.array([344,122,555,234,455,345])
f1={'family':'serif','color':'blue','size':23}
f2={'family':'serif','color':'pink','size':23}
plt.title('girl&boy',f1)
plt.xlabel('girl',f2)
plt.ylabel('boy',f2)
plt.plot(x,y)
plt.show()
```

```python
#adding grid
x=np.array([121,232,111,455,100,200])
y=np.array([344,122,555,234,455,345])
f1={'family':'serif','color':'blue','size':23}
f2={'family':'serif','color':'pink','size':23}
plt.title('girl&boy',f1)
plt.xlabel('girl',f2)
plt.ylabel('boy',f2)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
plt.pie(a,labels=y)
plt.show()
```

```python
#start angle
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
plt.pie(a,labels=y,startangle=180)
plt.show()
```

```python
#start angle
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
e=[0.4,0,0,0]
plt.pie(a,labels=y,explode=e)
plt.show()
```

```python
#shadow
#start angle
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
plt.pie(a,labels=y,shadow=True)
plt.show()
```

```python
#legend
#start angle
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
plt.pie(a,labels=y)
plt.legend()
plt.show()
```

```python
#legend title
a=np.array([3,12,56,78])
y=np.array(['apple','mango','kiwi','banana'])
plt.pie(a,labels=y)
plt.legend(title='fruits')
plt.show()
```
