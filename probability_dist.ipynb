probability for even and odd num

```python
import random
```

```python
num=[]
for i in range(100):
  num.append(random.randint(1,100))
  c=0
  for i in num:
    if i%2==0:
     c+=1
pro=c/len(num)
print(pro)
```

```python
print('probaboility for even num is ',pro*100,'%')
print('probability for odd num is',(1-pro)*100,'%')
```
