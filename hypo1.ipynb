confidence interval

```python
import numpy as np
from scipy.stats import norm
```

```python
d=[23,45,77,33,22,87,34,56]
```

```python
mean=np.mean(d)
st_e=np.std(d)/np.sqrt(len(d))
```

```python
print(mean)
print(st_e)
```

```python
c=0.95
z_score=norm.ppf((1+c)/2)
```

```python
u_b= mean+z_score*st_e
l_b=mean-z_score*st_e
```

```python
print('confidence interval',l_b,'to',u_b)
```

z-test

```python
x=23
mu=56
sigma=34
n=100
alpha=0.05
```

```python
z_score=((x-mu)/(sigma/np.sqrt(n)))
```

```python
p_val=2*(1-norm.cdf(abs(z_score)))
```

```python
print('z_score',z_score)
print('p_value',p_val)
```

```python
if p_val<alpha:
  print('reject H0')
else:
  print('fail to rejrct H0')
```

t-test


```python
import scipy.stats as stats
```

```python
x=[23,56,12,34,78]
mu=32
alpha=0.05
s=10
n=90
```

```python
t_stat,p_val=stats.ttest_1samp(x,mu)
```

```python
t_stat,p_val
```

```python
if p_val<alpha:
  print('reject H0')
else:
  print('fail to reject H0')
```
